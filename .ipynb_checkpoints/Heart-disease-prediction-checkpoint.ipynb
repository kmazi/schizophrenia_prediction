```python
import pandas as pd
import numpy as np
from keras import Model
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers
from keras.layers import Dropout
import matplotlib
import matplotlib.pyplot as plt
from sklearn import model_selection
from keras.utils.np_utils import to_categorical
import seaborn as sns
import math
```

```python
# Load dataset
dataset = pd.read_csv("Schizhophrenia_latest.csv")
```

```python
# Preprocessing data
# display first 5 rows
dataset.head()
```

```python
# describe stat info about dataset
dataset.info()
```

```python
dataset = dataset.replace(" ", None).replace("", None)
dataset.loc[:,"level_of_education"] = dataset["level_of_education"].str.strip().replace("", None)
dataset.loc[:,"Duration_of_Symtoms(months)"] = dataset["Duration_of_Symtoms(months)"].str.strip().replace("", None)
dataset.loc[:,"EXCESS_WRITING"] = dataset["EXCESS_WRITING"].str.strip().replace("", None)
dataset.loc[:,"SEEING_THINGS"] = dataset["SEEING_THINGS"].str.strip().replace("", None)
dataset.loc[:,"COHERENT"] = dataset["COHERENT"].str.strip().replace("", None)
dataset.loc[:,"NORMAL_VOLUME_AND_TUNE"] = dataset["NORMAL_VOLUME_AND_TUNE"].str.strip().replace("", None)
```

```python
# Drop hospital_number column
dataset = dataset.drop(["Hospital_Number","S/N"], 1)
```

```python
dataset.loc[:,'Age'] = dataset.Age.fillna(math.ceil(dataset.Age.mean())).astype('int32')

dataset.loc[:,'Gender'] = dataset.Gender.fillna(dataset.Age[0])

dataset.loc[:,'level_of_education'] = dataset.level_of_education.fillna(2)

dataset.loc[:,'Duration_of_Symtoms(months)'] = dataset.loc[:,'Duration_of_Symtoms(months)'].fillna(method='pad')
dataset.loc[:,'Duration_of_Symtoms(months)']

dataset.loc[:,'EXCESS_WRITING'] = dataset.EXCESS_WRITING.fillna(1)
```

```python
dataset = dataset.loc[~dataset['Diagnosis'].isna()]
```

```python
dataset = dataset.fillna(0)
dataset.info()
```

```python
dataset.loc[:, ['level_of_education', 'Duration_of_Symtoms(months)', 'EXCESS_WRITING',
               'SEEING_THINGS', 'COHERENT', 'NORMAL_VOLUME_AND_TUNE']].info()
```

```python
dataset.loc[:, ['level_of_education', 'Duration_of_Symtoms(months)', 'EXCESS_WRITING',
               'SEEING_THINGS', 'COHERENT', 'NORMAL_VOLUME_AND_TUNE']] = dataset.loc[:, ['level_of_education', 'Duration_of_Symtoms(months)', 'EXCESS_WRITING',
               'SEEING_THINGS', 'COHERENT', 'NORMAL_VOLUME_AND_TUNE']].astype('float32')
```

```python
# Get trainset and test set
dataset.loc[:,'Diagnosis'] -= 1
X = np.array(dataset.drop(['Diagnosis'], 1))
Y = np.array(dataset['Diagnosis'])
```

```python
X
```

```python
Y
```

```python
# normalize input
mean = X.mean(axis=0)
X -= mean
std = X.std(axis=0)
X /= std
```

```python
X.shape
```

```python
# create model
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, stratify=Y, random_state=42, test_size=0.2)
```

```python
# make y_train and y_test categorical
Y_train = to_categorical(Y_train)
Y_test = to_categorical(Y_test)
```

```python
# create keras model
def create_model():
    """create ai model with keras"""
    model = Sequential()
    # layer 1
    model.add(
        Dense(16, input_dim=71, kernel_initializer="normal", kernel_regularizer=regularizers.l2(0.01), activation='relu')
    )
    # layer 2
    model.add(
        Dense(8, kernel_initializer="normal", kernel_regularizer=regularizers.l2(0.01), activation='relu')
    )
    model.add(Dropout(0.25))
    # output layer
    model.add(
        Dense(2, activation="softmax")
    )
    
    model.compile(loss="categorical_crossentropy", optimizer="rmsprop", metrics=["accuracy"])
    return model
```

```python
model = create_model()
model.summary()
```

```python
# Fit model to training data
training = model.fit(X_train, Y_train, validation_data=(X_test, Y_test), epoch=50, batch_size=10)
```
