```python
import pandas as pd
import numpy as np
from keras import Model
import matplotlib
import matplotlib.pyplot as plt
from sklearn import model_selection
import seaborn as sns
import math
```

```python
# Load dataset
dataset = pd.read_csv("Schizhophrenia_latest.csv")
```

```python
# Preprocessing data
# display first 5 rows
dataset.head()
```

```python
# describe stat info about dataset
dataset.info()
```

```python
datasets = dataset.replace(" ", None)
datasets.loc[:,"level_of_education"] = datasets["level_of_education"].str.strip().replace("", None)
```

```python
# Drop hospital_number column
datasets = datasets.drop(["Hospital_Number"], 1)
```

```python
dataset.loc[:,'Age'] = dataset.Age.fillna(math.ceil(dataset.Age.mean())).astype('int32')
```

```python
dataset.loc[:,'Gender'] = dataset.Gender.fillna(dataset.Age[0])
```

```python
dataset.loc[:,'level_of_education'] = dataset.level_of_education.fillna(2)
```

```python
dataset.loc[:,'Duration_of_Symtoms(months)'] = dataset.loc[:,'Duration_of_Symtoms(months)'].str.replace(' ', '1')
dataset.loc[:,'Duration_of_Symtoms(months)'] = dataset.loc[:,'Duration_of_Symtoms(months)'].fillna(method='pad')
dataset.loc[:,'Duration_of_Symtoms(months)']
```

```python
dataset.loc[:,'EXCESS_WRITING'] = dataset.loc[:,'EXCESS_WRITING'].str.strip().str.replace(' ', '1')
dataset.loc[:,'EXCESS_WRITING'] = dataset.EXCESS_WRITING.fillna(1)
```

```python
dataset.loc[:,'SEEING_THINGS'] = dataset.loc[:,'SEEING_THINGS'].str.replace(' ', '1')
```

```python
dataset = dataset.loc[~dataset['Diagnosis'].isna()]
```

```python
dataset = dataset.fillna(0)
dataset = dataset.drop(['S/N'], 1)
dataset
```
